```python
# House Predication
House["Area Type"] = House["Area Type"].map({"Super Area": 1,
                                           "Carpet Area": 2,
                                           "Built Area": 3})
House["City"] = House["City"].map({"Mumbai": 4000, "Chennai": 6000,
                                 "Bangalore": 5600, "Hyderabad": 5000,
                                 "Delhi": 1100, "Kolkata": 7000})
House["Furnishing Status"] = House["Furnishing Status"].map({"Unfurnished": 0,
                                                           "Semi-Furnished": 1,
                                                           "Furnished": 2})
House["Tenant Preferred"] = House["Tenant Preferred"].map({"Bachelors/Family": 2,
                                                         "Bachelors": 1,
                                                         "Family": 3})
print(House.head())
```

```python
 # Splitting data
from sklearn.model_selection import train_test_split
x = np.array(House[["BHK", "Size", "Area Type", "City",
                    "Furnishing Status", "Tenant Preferred",
                    "Bathroom"]])
y = np.array(House[["Rent"]])
X_train, X_test, Y_train, Y_test  = train_test_split(x,y, test_size=0.10, random_state = 42)
```

```python
# Now let’s train a house rent prediction model using an LSTM neural network model:
from keras.models import Sequential
from keras.layers import Dense, LSTM
model = Sequential()
model.add(LSTM(128, return_sequences=True,
              input_shape=(X_train.shape[1],1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, Y_train,batch_size=1, epochs=21)
```

```python
# Predict the rent of a House using Trained Model
print("Enter House Details to Predict Rent")
a = int(input("Number of BHK:"))
b = int(input("Size of the House"))
c = int(input("Area Type (Super Area = 1, Carpet Area = 2, Built Area = 3):"))
d = int(input("Pin Code of the City"))
e = int(input("Furnishing Status of the House (Unfurished =0, semi-Furnished = 1, Furished = 2):"))
f = int(input("Tenant Type (Bachelors =1, Bachelors/Family = 2, only Family = 3):"))
g = int (input("Number of bathrooms:"))
features = np.array([[a,b,c,d,e,f,g]])
print("Prediction House Price = ",model.predict(features))
```
