```python
import plotly.express as px
fig = px.line(Delhi, x="date",
                 y="meantemp",
                 title= "meantemp of Delhi")
fig.show()
```

```python
fig = px.line(Delhi, x="date",
                 y="humidity",
                 title="Humidity of delhi")
fig.show()
```

```python
fig = px.line(Delhi, x="date",
              y="wind_speed",
              title="wind_speed of delhi")
fig.show()
```

```python
fig = px.scatter(Delhi, x="humidity",
                 y="meantemp",
                 trendline="ols",
                 title= "Relationship b/w humidity nd meantemp")
fig.show()
```
